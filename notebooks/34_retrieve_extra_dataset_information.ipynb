# Preparação do ambiente

## Bibliotecas

```python
import pandas as pd
```

## Constantes e funções auxiliares

```python
# arquivos utilizados na rotulagem
# alguns nomes de arquivos foram alterados em relação à execução original para 
# conferir melhor coerência e entendimento

# arquivos brutos do Regulatron
FILE_AMAZON = '../datasets/rotulagem_consolidada/dados_brutos_regulatron/amazon.parquet.gzip'
FILE_MAGALU = '../datasets/rotulagem_consolidada/dados_brutos_regulatron/magalu.parquet.gzip'
FILE_ML = '../datasets/rotulagem_consolidada/dados_brutos_regulatron/ml.parquet.gzip'

# arquivos de anúncios preparados e amostra para rotulagem
FILE_ANUNCIOS = '../datasets/rotulagem_consolidada/dados_tratados_regulatron/anuncios.parquet'
FILE_ANUNCIOS_AMOSTRA = '../datasets/rotulagem_consolidada/dados_tratados_regulatron/anuncios_amostra_fiscaliza.parquet'

# arquivo de tarefas para o fiscaliza
FILE_TAREFAS_FISCALIZA = '../datasets/rotulagem_consolidada/tarefas_fiscaliza/anuncios_tarefas_fiscaliza.parquet'
FILE_TAREFAS_FISCALIZA_FECHADAS = '../datasets/rotulagem_consolidada/tarefas_fiscaliza/anuncios_tarefas_fiscaliza_fechadas.parquet'
```

# Carga e prepação dos dados

```python
%%time
# todos os anúncios raspados pelo Regulatron
# um dataframe para cada pois há colunas diferentes
df_amazon = pd.read_parquet(FILE_AMAZON)
df_magalu = pd.read_parquet(FILE_MAGALU)
df_ml = pd.read_parquet(FILE_ML)
```

```python
%%time
# anúncios da amostra utilizada para rotulagem
df_anuncios_amostra = pd.read_parquet(FILE_ANUNCIOS_AMOSTRA)
df_anuncios_amostra
```

# Análise

```python
df_categoria = None
for df in (df_amazon,df_magalu,df_ml):    
    columns_to_keep = [col for col in df.columns if 'cat' in col.lower()]
    columns_to_keep.sort()
    columns_to_keep = ['Link']+columns_to_keep
    
    if df_categoria is None:
        df_categoria = df[columns_to_keep]
    else:
        df_categoria = pd.concat([df_categoria, df[columns_to_keep]])

df_categoria = df_categoria.reset_index(drop=True)
df_categoria
```

```python
df_anuncios_amostra = df_anuncios_amostra.merge(df_categoria,left_on='link',right_on='Link')
df_anuncios_amostra.Categoria.value_counts()
```
