# Carga e prepação dos dados

## Reindexar

Reindexar dados da Amazon. Os dados reindexados serão usados como base para o desenvolvimento de experimentos de classificação.

```python
file_amazon_raw = '../datasets/amazon/raw/amazon.parquet.gzip'
file_amazon_reindexed = '../datasets/amazon/clean/amazon_reindexed.parquet'

if osp.exists(file_amazon_reindexed):
    print(f'Read existing file {file_amazon_reindexed}.')
    df_amazon = pd.read_parquet(file_amazon_reindexed)
    
else:
    print(f'Read existing file {file_amazon_raw}, redindex and save to new file {file_amazon_reindexed}')
    df_amazon = pd.read_parquet(file_amazon_raw)
    # fill na values with NULL_STRING
    df_amazon = df_amazon.fillna(NULL_STRING)
    # reindex
    df_amazon = df_amazon.reset_index(drop=True)
    # save to new file
    df_amazon.to_parquet(file_amazon_reindexed,compression='gzip')
df_amazon
```

```python
columns_to_keep = ['cat_0','cat_1','cat_2','cat_3','cat_4','cat_5']
df_amazon['cat_full']=df_amazon[columns_to_keep].apply(lambda row: '_'.join(row),axis=1)
df_amazon['cat_full'].value_counts()
```

```python
df_cat_count = df_amazon.cat_0.value_counts()
df_cat_count
```

```python
file_cat_to_label = '../datasets/amazon/targets/amazon_cat_to_label.xlsx'

columns_to_keep = ['cat_0','cat_1','cat_2','cat_3','cat_4','cat_5','cat_full','Nome']
df_cat = df_amazon[df_amazon['cat_0']!=NULL_STRING][columns_to_keep]
df_cat = df_cat.groupby(columns_to_keep[:-1],as_index=False).count().sort_values(by='Nome',ascending=False)

# exportar em xlsx para manipular os dados manualmente
# caso já exista, exportar em formato csv para evitar perda dos dados já manipulados
if osp.exists(file_cat_to_label):
    file_cat_to_label_csv = file_cat_to_label.replace('xlsx', 'csv')
    print(f'Alert! File {file_cat_to_label} exists. Save to file{file_cat_to_label_csv}')
    df_cat.to_csv(file_cat_to_label_csv,index=False,sep=';')
else:    
    print(f'Save to file{file_cat_to_label}')
    df_cat.to_excel(file_cat_to_label,index=False)
df_cat
```

```python
columns_to_keep = columns_to_keep = ['cat_0','cat_1','cat_2','cat_3','cat_4','cat_5','cat_full','Link','Nome']
df_cat_sample = df_amazon[columns_to_keep]

# map cat with less than 1000 items
map_cat_outros = {cat:'Outros' for cat in df_cat_count[df_cat_count<1000].index}
df_cat_sample.loc[:,'cat_full'] = df_cat_sample[['cat_0','cat_full']].apply(lambda cat: map_cat_outros[cat['cat_0']] if cat['cat_0'] in map_cat_outros.keys() else cat['cat_full'],axis=1)

df_cat_sample = df_cat_sample.groupby('cat_full').sample(10,replace=True,random_state=22000).drop_duplicates()
file_cat_sample = '../datasets/cat_to_label_sample.xlsx'
if osp.exists(file_cat_sample):
    file_cat_sample_csv = file_cat_sample.replace('xlsx','csv')
    print(f'Alert! File {file_cat_sample} exists. Save to file {file_cat_sample_csv}')
    df_cat_sample.to_csv(file_cat_sample_csv,index=False,sep=';')
else:    
    print(f'Save to file {file_cat_sample}')
    df_cat_sample.to_excel(file_cat_sample,index=False)
df_cat_sample
```
